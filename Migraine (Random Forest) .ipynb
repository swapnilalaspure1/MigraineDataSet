```python
# Creating User Define Function To Replace Null Values In Data Frame..!
import pandas as pd

def replace_null_values(df):
    for column in df.columns:
        if df[column].isnull().any():
            df[column].fillna(df[column].mode().iloc[0], inplace=True)
    return df
```

```python
# Importing library 
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Applying Select K Best class to extract top 10 best features
best_features = SelectKBest(score_func=chi2,
                            k=10)
fit = best_features.fit(X,
                        Y)

# Now Extracting Scores
df_score = pd.DataFrame(fit.scores_)
df_columns = pd.DataFrame(X.columns)

# Now Concate two df for better visualization
feature_score = pd.concat([df_score,
                           df_columns],
                           axis=1)
feature_score.columns = ["Feature Score", 
                         "Features"]
```

```python
# Now we are droping those two columns
mgr = mgr.drop(columns=['Dysarthria', 'Ataxia','Nausea'])
mgr.head()
```

```python
# Shape of data set
mgr.shape
```

```python
# Now We Defining X and Y
X = mgr.drop(columns = ['Type'])
Y = mgr[['Type']]
```

```python
# Spliting the data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,
                                                 Y,
                                                 test_size=0.2,
                                                 random_state=42)

len(X_train),len(X_test),len(Y_train),len(Y_test)
```

```python
# Building a model
from sklearn.ensemble import RandomForestClassifier as RFC
rfc_model = RFC(n_estimators=1000,
                random_state=42,
                n_jobs=-1)
# Now fit the model on train and test data 
rfc = rfc_model.fit(X_train,Y_train)
rfc
```

```python
# Prediction
Y_test['Predicted_Type'] = rfc_model.predict(X_test)
Y_test
```

### Checking Classification Report and Accuraccy Score

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(Y_test['Type'],Y_test['Predicted_Type']))
```

```python
# Accuraccy Score
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test['Type'],Y_test['Predicted_Type']))
```

#### After Seeing Classification report we can clearly see that, the recall values for category 1 and 4 very low compaired to other category.
#### Now we are going to apply SMOTE technique

```python
# lets find the number of datapoints in training dataset for target variable
print('count of 0 in Y_train is:', sum(Y_train['Type']==0))
print('count of 1 in Y_train is:', sum(Y_train['Type']==1))
print('count of 2 in Y_train is:', sum(Y_train['Type']==2))
print('count of 3 in Y_train is:', sum(Y_train['Type']==3))
print('count of 4 in Y_train is:', sum(Y_train['Type']==4))
print('count of 5 in Y_train is:', sum(Y_train['Type']==5))
print('count of 6 in Y_train is:', sum(Y_train['Type']==6))
```

```python
# Applying Smote technique
from imblearn.over_sampling import SMOTE as SMT
smote = SMT(n_jobs=-1,
            random_state=42)

# We will fit the smote object on X_train and Y_tarin
X_train_new, Y_train_new = smote.fit_resample(X_train, Y_train)
```

```python
# lets find the number of datapoints in new training dataset for target variable
print('count of 0 in Y_train_new is:', sum(Y_train_new['Type']==0))
print('count of 1 in Y_train_new is:', sum(Y_train_new['Type']==1))
print('count of 2 in Y_train_new is:', sum(Y_train_new['Type']==2))
print('count of 3 in Y_train_new is:', sum(Y_train_new['Type']==3))
print('count of 4 in Y_train_new is:', sum(Y_train_new['Type']==4))
print('count of 5 in Y_train_new is:', sum(Y_train_new['Type']==5))
print('count of 6 in Y_train_new is:', sum(Y_train_new['Type']==6))
```

```python
# Now we will again run the model with new x and Y train 
# Building a model
from sklearn.ensemble import RandomForestClassifier as RFC
rfc_model = RFC(n_estimators=1000,
                random_state=42,
                n_jobs=-1)
# Now fit the model on train and test data 
rfc = rfc_model.fit(X_train_new,Y_train_new)
rfc
```

```python
# Prediction after applying smote
Y_test['Predicted_Type'] = rfc_model.predict(X_test)
Y_test
```

# Report and Cross Validation

```python
# Classification Report 
from sklearn.metrics import classification_report
print(classification_report(Y_test['Type'],Y_test['Predicted_Type']))
```

```python
# Accuraccy Score
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test['Type'],Y_test['Predicted_Type']))
```

```python
# Cross Validation report 
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator= rfc, 
                             X= X_train_new, 
                             y= Y_train_new,
                             cv= 5)
accuracies
```
